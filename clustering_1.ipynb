```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
#performing hierarchical Clustering
crime = pd.read_csv("crime_data.csv")
```

```python
crime
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df_norm = norm_func(crime.iloc[:,1:])
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm, method='complete'))
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'complete')
```

```python
y_hc = hc.fit_predict(df_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
crime['h_clusterid'] = hc.labels_
```

```python
crime
```

```python
#performing K-Means clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime_df = scaler.fit_transform(crime.iloc[:,1:])
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters_new = KMeans(4, random_state=42)
clusters_new.fit(scaled_crime_df)
```

```python
clusters_new.labels_
```

```python
crime['clusterid_new'] = clusters_new.labels_
```

```python
clusters_new.cluster_centers_
```

```python
crime.groupby('clusterid_new').agg(['mean']).reset_index()
```

```python
#performing DBSCAN
from sklearn.cluster import DBSCAN
```

```python
crime = pd.read_csv("crime_data.csv")
```

```python
crime
```

```python
crime.drop(['City'],axis=1,inplace=True)
```

```python
array = crime.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
dbscan = DBSCAN(eps=2, min_samples=4)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([crime,cl],axis=1)
```
