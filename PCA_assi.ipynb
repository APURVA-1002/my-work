```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
#PCA
wine = pd.read_csv("wine.csv")
wine.describe()
wine.head()
```

```python
Wine = wine.values
```

```python
Wine
```

```python
wine_normal = scale(Wine)
wine_normal
```

```python
pca = PCA()
pca_values = pca.fit_transform(wine_normal)
```

```python
pca_values
```

```python
var = pca.explained_variance_ratio_
var
```

```python
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
plt.plot(var1,color="red")
```

```python
pca_values[:,0:1]
```

```python
x = pca_values[:,0:1]
y = pca_values[:,1:2]
plt.scatter(x,y)
```

```python
finalDf = pd.concat([pd.DataFrame(pca_values[:,0:2],columns=['pc1','pc2']), wine[['Type']]], axis = 1)
```

```python
import seaborn as sns
sns.scatterplot(data=finalDf,x='pc1',y='pc2',hue='Type')
```

```python
#now performing hierarchical clustering
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
new_wine = pd.concat([pd.DataFrame(pca_values[:,0:3],columns=['pc1','pc2','pc3']), wine[['Type']]], axis = 1)
new_wine
```

```python
dendrogram = sch.dendrogram(sch.linkage(new_wine, method='average'))
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'average')
```

```python
y_hc = hc.fit_predict(new_wine)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
new_wine['h_clusterid'] = hc.labels_
```

```python
new_wine
```

```python
#now performing K-Means clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(new_wine)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clusters_new = KMeans(4, random_state=42)
clusters_new.fit(new_wine)
```

```python
clusters_new.labels_
```

```python
new_wine['clusterid_new'] = clusters_new.labels_
```

```python
clusters_new.cluster_centers_
```

```python
new_wine.groupby('clusterid_new').agg(['mean']).reset_index()
```
